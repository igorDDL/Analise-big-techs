```python
!pip install yfinance --upgrade --no-cache-dir

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print("Ambiente configurado")
```

```python
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN']

dados = yf.download(tickers, start="2020-01-01")['Close']

print(dados.head())
```

```python
#Evolução dos Preços
dados.plot()
plt.title('Evolução do Preço das Ações (2020 - Hoje)')
plt.ylabel('Preço ($)')
plt.xlabel('Data')
plt.show()

#Retorno Diário
retornos = dados.pct_change()

plt.figure(figsize=(10,5))
sns.histplot(retornos['AAPL'].dropna(), bins=100, color='purple', kde=True)
plt.title('Distribuição dos Retornos Diários da Apple')
plt.xlabel('Retorno Diário')
plt.show()

#Mapa de Calor
plt.figure(figsize=(8,6))
sns.heatmap(retornos.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlação entre as Ações')
plt.show()
```

#  Conclusão e Insights da Análise

Neste projeto, analisamos o comportamento das ações de quatro grandes empresas de tecnologia (**Apple, Microsoft, Google e Amazon**) no período de 2020 até o presente. A utilização de bibliotecas como `yfinance`, `pandas` e `seaborn` permitiu extrair, processar e visualizar dados históricos para identificar padrões econômicos.

# Principais Descobertas:

1.  **Tendência e Contexto Macroeconômico:**
    * Observamos uma tendência geral de alta (bull market), impulsionada inicialmente pela digitalização forçada durante a pandemia e, mais recentemente (2023-2024), pelo *hype* da Inteligência Artificial Generativa.
    * As quedas observadas coincidem com períodos de aumento da taxa de juros global, demonstrando a sensibilidade do setor de tecnologia ao custo do capital.

2.  **Comportamento Estatístico (Risco):**
    * A análise do histograma de retornos diários revelou uma distribuição próxima da normal (Curva de Gauss), com a maior frequência de variações próxima a 0%.
    * Isso indica que, apesar da percepção de volatilidade, os movimentos extremos (caudas da distribuição) são eventos menos frequentes do que a oscilação natural do mercado.

3.  **Correlação Setorial:**
    * O *Heatmap* de correlação evidenciou coeficientes altos (geralmente acima de 0.6) entre os ativos.
    * **Conclusão Estratégica:** Isso comprova que essas empresas "andam juntas". Para um investidor, manter uma carteira composta apenas por essas *Big Techs* oferece pouca proteção (diversificação), pois um evento negativo no setor afetaria todo o portfólio simultaneamente.
